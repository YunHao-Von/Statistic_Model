```python
#读取邻接矩阵
import numpy as np
connect_matrix=np.load('connect_matrix.npy')
weighted_edges=[]
for i in range(connect_matrix.shape[0]):
    for j in range(len(connect_matrix[0])):
        if connect_matrix[i,j]!=0:
            weighted_edges.append((i,j,connect_matrix[i,j]))
```

```python
import networkx as nx
DG = nx.DiGraph()
DG.add_nodes_from(list(range(3412)))
DG.add_weighted_edges_from(weighted_edges)
```

```python
in_degree=[]
num_nodes = DG.number_of_nodes()
nodes = DG.nodes()
for i in nodes:
    in_degree.append(DG.in_degree(i))
DG.add_node(3412)
for node in nodes:
    if node!=3412:
        DG.add_edge(3412, node)
        DG.add_edge(node,3412)
```

```python
new_connect_matrix=np.zeros((3413,3413))
new_connect_matrix[:3412,:3412]=connect_matrix
new_connect_matrix[:,3412]=np.ones_like((3413,))
new_connect_matrix[-1,:3412]=np.array(in_degree)+1
```

```python
LR = dict.fromkeys(nodes, 1.0)
LR[3412]=0.0
```

```python
for i in range(new_connect_matrix.shape[0]):
    temp_sum=np.sum(new_connect_matrix[i,:])
    for j in range(len(new_connect_matrix[i,:])):
        new_connect_matrix[i,j]=new_connect_matrix[i,j]/temp_sum
```

```python
all_error=[]
while True:
    tempLR = {}
    for node1 in DG.nodes():
        s = 0.0
        for node2 in list(DG.predecessors(node1)):
            s += LR[node2]*new_connect_matrix[node2,node1]
        tempLR[node1] = s
        # 终止条件:LR值不在变化
    error = 0.0
    for n in tempLR.keys():
        error += abs(tempLR[n] - LR[n])
    all_error.append(error)
    print(error)
    if error <= 0.0000001:
        break
    LR = tempLR
avg = LR[3412] / num_nodes
LR.pop(3412)
for k in LR.keys():
    LR[k] += avg
```
