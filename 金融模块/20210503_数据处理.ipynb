```python
import tushare as ts
import pandas as pd
import time
pro=ts.pro_api()
stock_list = pd.read_csv('股票列表.csv',encoding='utf-8')
for i in range(3880,stock_list.shape[0]):
    temp_code=stock_list.iloc[i,0]
    temp_start_date=stock_list.iloc[i,5]
    end_date='20210425'
    temp_result=pro.daily(ts_code=temp_code, start_date=str(temp_start_date), end_date=end_date)
    temp_result.to_csv('处理后数据/日线数据/'+str(temp_code[0:6])+'.csv',index=False)
    time.sleep(2)
    print(i)
```
